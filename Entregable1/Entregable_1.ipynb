```python
sales = sales.rename({'ITEM_CODE' : 'ITEM'}, axis = 'columns')  #Cambio de Nombre a Columna
sales
```

# Unión de Dataframes

```python
df_category_segment = pd.merge(category, segment, on = 'ID_CATEGORY', how = 'inner') # Unión de los dataframes Category & Segment
df_category_segment
```

```python
df_sales_product = pd.merge(sales, product, on = 'ITEM', how = 'inner')  # Unión de los dataframes Sales & Product
df_sales_product = df_sales_product.rename({'CATEGORY' : 'ID_CATEGORY'}, axis = 'columns') # Cambio de nombre a columna
df_sales_product
```

```python
# Unión de Ventas_Producto con Categoría_Segmento

df_sales_product_category_segment = pd.merge(df_sales_product, df_category_segment, on = ['ID_CATEGORY', 'FORMAT', 'ATTR1', 'ATTR2', 'ATTR3'], how = 'inner')
df_sales_product_category_segment
```

```python
df_sales_product_category_segment_calendar = pd.merge(calendar, df_sales_product_category_segment, on = 'WEEK', how = 'right') # Unión con Calendario
df_sales_product_category_segment_calendar
```

```python
df_final = df_sales_product_category_segment_calendar.groupby(['DATE', 'ITEM', 'SEGMENT', 'REGION', 'BRAND', 'ITEM_DESCRIPTION', 'FORMAT']).agg({
    'TOTAL_UNIT_SALES' : 'sum',
    'TOTAL_VALUE_SALES' : 'sum',
    'TOTAL_UNIT_AVG_WEEKLY_SALES' : 'mean'
}).reset_index()  # Agrupación final

df_final.info()
```

```python
df_final.to_csv('datos_filtrados.csv')  #Exportación de DataFrame a .csv
```

```python
df_attr = df_sales_product_category_segment_calendar.groupby(['DATE', 'ITEM', 'CATEGORY', 'SEGMENT', 'REGION', 'BRAND', 'FORMAT', 'ATTR1', 'ATTR2']).agg({
    'TOTAL_UNIT_SALES' : 'sum',
    'TOTAL_VALUE_SALES' : 'sum',
    'TOTAL_UNIT_AVG_WEEKLY_SALES' : 'mean'
}).reset_index()  # Agrupación con attr

df_attr.info()
```

```python
df_attr.to_csv('datos_con_attr.csv')  #Exportación de DataFrame a .csv
```

```python
df_final_WEEK = df_sales_product_category_segment_calendar.groupby(['WEEK', 'ITEM', 'CATEGORY', 'SEGMENT', 'REGION', 'BRAND', 'ITEM_DESCRIPTION']).agg({
    'TOTAL_UNIT_SALES' : 'sum',
    'TOTAL_VALUE_SALES' : 'sum',
    'TOTAL_UNIT_AVG_WEEKLY_SALES' : 'mean'
}).reset_index()  # Agrupación final

df_final_WEEK.info()
```

```python
df_final_WEEK.to_csv('datos_con_WEEK.csv')  #Exportación de DataFrame a .csv
```
